### 2) Labelencoding'de ilgili kolon için ölçeklendirme nasıl yapılır? (Verinin doğru etkisiyle sayısal olarak dönüştürülmesi)

Label Encoding’de, her bir veri için **alfabetik** sıralamaya göre benzersiz bir tam sayı atanır. Örneğin cinsiyet değişkeninde eşsiz iki değer vardır : Female ve Male. F alfabetik olarak daha önde olduğu için Female 0, Male 1 değerini alır.

Eğitim değişkenine ait veriler **Ordinal Kategoriktir**. Yani aralarında hiyerarşik bir sıra vardır. PhD > Master > Graduate > High School > Secondary School > Pre-School diyebiliriz. bu eğitim durumları arasında fark olduğunu belirtmek isteyebiliriz. Bu durumda 0'dan 5'e sayılarla temsil etmek mantıklı olabilir. Unvan, eğitim durumu, rütbe ordinal kategorik değişkenlere örnek olabilir.

Kısaca aşağıdaki durumlarda **Label Encoding** uygulanır:
* Kategorik özellik sıralıdır
* Tek etkin kodlama yüksek bellek tüketimine yol açabileceğinden kategori sayısı oldukça fazladır.

Label Encoding ile ilgili olası bir sorun, etiket kodlaması bir ilişki sunarken, çoğu zaman kategoriler arasında herhangi bir ilişki olmamasıdır. Orjinal verilerde sayısal bir ilişki olmamasına rağmen sayısal değerler kullanılarak kodlanmış veriler arasında sayıyla ilgili bir bağlantı kurulabilir. Bu sorunun üstesinden gelmek için One Hot Encoding kullanılabilir.

### 3) Imbalance datasette train test split yaparken neleri göz önünde bulundurmalıyız?

Imbalanced datasetler, aşağıdakiler de dahil olmak üzere birçok alanda görünür:
* Fraud detection
* Spam filtering
* Disease screening
* SaaS subscription churn
* Advertising click-throughs

Imbalanced datasetler modeli eğitirken yanlılık sorununun ortaya çıkarır. Modele çarpık bir dünya görüşü gösterilirse, gelecekle ilgili tahminde bulunurken bu önyargıyı taşır.

Verileri eşitlemek için veri noktaları eklediğinde veya çıkarıldığında, bu işlemin sadece training data'ları için yapılması gerekiyor. Modele dünyanın eşit bir görünümü gösterilmek istenir. Bu nedenle ona eşit veriler verilir. Test data'ları gerçek dünyayı yansıtmalıdır.

Imbalanced datasetleri işlemek için gerekli iki tekniğe bakalım: 

**1. Up-sample**

Up-sample, az veriye sahip grubun sinyalini güçlendirmek için gözlemleri rastgele çoğaltma sürecidir.

Bunu yapmak için birkaç yöntem vardır. En yaygın yöntem, değiştirme ile  yeniden örneklemedir (resample with replacement).

**2. Down-sample**

Down-sample,   çoğunluk sınıfın öğrenme algoritmasında baskın olmasını önlemek için gözlemlerin rastgele çıkarılmasını içerir. Bunu yapmak için en yaygın  yöntem, değiştirmeden yeniden örneklemedir (resampling without replacement).

### 4) Validation dataseti (modelin görmediği) nasıl oluşturulur ve nasıl predict etmeye hazır hale getirilir?

Eğitim (train) setindeki sonuçları değerlendirmek için doğrulama seti (validation) kullanılır. Validation set, modelin üzerinde eğitim alınmamış ve hiperparametreleri ayarlamak için kullanılan veriler. Validation set ile her eğitim sonunda model etkinliğini test edip nihai modele karar verdikten sonra test seti ile test ederek modelimizin etkinliği daha sağlıklı bir şekilde gözlemleyebiliriz.


Konunun daha net anlaşılabilmesi için bir sınıflandırma modeli üzerinden açıklayalım. Elimizde kişilerin kanser olup olmadığını tahminlemeye çalıştığımız bir göğüs kanseri veri seti olsun. Veri setini 80%-20% oranında training ve test seti olarak ayıralım. Modelimizi train ettik sonra içerisine test verisinden bir örnek koyduk o da bize bu kişinin kanser hastası olduğunu söyledi ama gerçekte o kişi kanser hastası değil buna göre hatamıza baktık, parametrelerimizi değiştirip yeniden train ettik. Zamanla kendimizi bu test verisinden aldığımız hatalara göre adapte ediyoruz yani test verisine overfit ediyoruz. Bu sebeple validation set ekliyoruz, hataya bakıp parametre değiştirme işlemini validation set’te yapıyoruz ardından yeni çıkan modeli test verisiyle test ediyoruz böylece modelin gerçekten iyi bir performans sergileyip sergilemediğini görebiliyoruz.

* Sklarn üzerinden değişkenlerin ayrıştırılarak bölünür
* X input değişkenler Y ise hedef değişken ayrıştırılarak fonksiyona dahil edilir.
* Veri seti öncelikle train ve test olarak ayrılır.
* Train veri seti kendi içinden validation bölümünü ayırılır.

### Part 3: Makine Öğrenmesi Modelleri

```python
features = []
for col in df2.columns:
  if col != 'fraud_reported':
    features.append(col)

# Target - bağımlı değişken
target = 'fraud_reported'

X = df2[features] # bağımsız değişkenler
y = df2[target]   # target değişken
```

```python
X = df2[features]
```

```python
X
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0, stratify=y)
```

Ölçekleme (scaling) özelliği değişkenleri:

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler  = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
import sklearn.metrics as skm
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()

knn.fit(X_train, y_train)
```

```python
preds = knn.predict(X_test)
```

```python
print("First draft KNN results:")
print("Accuracy:  " + str(round(skm.accuracy_score(y_true=y_test,y_pred=preds) * 100, 2)) + "%")
print("Precision: " + str(round(skm.precision_score(y_true=y_test,y_pred=preds,average='weighted') * 100, 2)) + "%")
print("Recall:    " + str(round(skm.recall_score(y_true=y_test,y_pred=preds,average='weighted') * 100, 2)) + "%")
print("F1-Score:  " + str(round(skm.f1_score(y_true=y_test,y_pred=preds,average='weighted') * 100, 2)) + "%")
```
